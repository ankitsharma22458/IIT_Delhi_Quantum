```python
import pennylane as qml
from pennylane import numpy as np

# define the device
dev = qml.device('default.qubit', wires=3)

# define cost hamiltonian
@qml.qnode(dev, interface='autograd')
def cost_hamiltonian(weights):
    qml.RX(weights[0], wires=0)
    qml.RY(weights[1], wires=1)
    qml.RZ(weights[2], wires=2)
    return qml.expval(qml.Hermitian(np.array([[1, 0], [0, -1]]), wires=[0, 1]))

# define the cost function
@qml.qnode(dev, interface='autograd')
def cost(weights):
    return cost_hamiltonian(weights)

# define the travel time
@qml.qnode(dev, interface='autograd')
def travel_time(weights):
    return qml.expval(qml.Hadamard(wires=2))

# define the total cost
def total_cost(weights):
    return cost(weights) + travel_time(weights)

# define the optimizer
opt = qml.AdamOptimizer(0.05, beta1=0.9, beta2=0.999)

# run the optimization
steps = 10
weights = np.array([0.0, 0.0, 0.0])

for i in range(steps):
    weights = opt.step(total_cost, weights)

print(weights)
```
